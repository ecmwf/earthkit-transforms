# Calculate the Area Mean of Gridded Data over a Geometries

This notebook demonstrates how to compute the area mean of some sample 2m temperature in ERA5 data for a given geometry (such as an administrative region) using earthkit-data and earthkit-transforms. This is useful for generating regional climate statistics.


```python
from earthkit import data as ekd
from earthkit import transforms as ekt
from earthkit.transforms._tools import earthkit_remote_test_data_file

remote_era5_file = earthkit_remote_test_data_file("era5-Europe-sfc-2m-temperature-3deg-2015-2017.grib")
era5_data = ekd.from_source("url", remote_era5_file)
ds = era5_data.to_xarray()

remote_geometry_url = earthkit_remote_test_data_file("NUTS_RG_60M_2021_4326_LEVL_0.geojson")
geometry_data = ekd.from_source("url", remote_geometry_url).to_geopandas()

# Calculate area mean
area_mean = ekt.spatial.reduce(ds, geometry_data, how="mean", weights="latitude")

area_mean
```

## Using a bounding box instead of a geometry

Instead of loading a geometry file, you can define the region of interest as a
simple bounding box using the `area` keyword argument. This is a dictionary with
keys `"north"`, `"south"`, `"east"`, and `"west"`.

```python
# Define a bounding box covering central Europe
area = {"north": 55, "south": 45, "east": 20, "west": 5}

# Calculate area mean using the bounding box
area_mean_bbox = ekt.spatial.reduce(ds, area=area, how="mean", weights="latitude")

area_mean_bbox
```
