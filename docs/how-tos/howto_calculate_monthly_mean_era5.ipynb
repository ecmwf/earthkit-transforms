# How to Calculate the Monthly Mean of ERA5 Data

This notebook demonstrates how to calculate the monthly mean 2m temperature for ERA5 data using earthkit-data and earthkit-transforms. Monthly means are common for climate summaries and reporting.


```python
from earthkit.data.utils.testing import earthkit_remote_test_data_file
from earthkit import data as ekd
from earthkit.transforms import temporal

remote_era5_file = earthkit_remote_test_data_file("era5-Europe-sfc-2m-temperature-3deg-2015-2017.grib")
era5_data = ekd.from_source("url", remote_era5_file)
ds = era5_data.to_xarray()

# Calculate monthly mean
monthly_mean = temporal.reduce(ds, how="mean", frequency="month")

monthly_mean
```
