```python
from earthkit import data as ekd
from earthkit.transforms._transform_inputs import transform_inputs_decorator
import xarray as xr
import numpy as np
```

```python
from earthkit.data.testing import earthkit_remote_test_data_file
ekd.settings.set("cache-policy", "user")
remote_era5_file = earthkit_remote_test_data_file("era5_temperature_europe_20150101.grib")
ekds = ekd.from_source("url", remote_era5_file)
ekds
```

```python
@transform_inputs_decorator()
def xrDataArray_function(data: xr.DataArray, astr: str=""):
    print(type(data))
    return data

this_da = xrDataArray_function(ekds)
```

```python
@transform_inputs_decorator()
def xrDataset_function(data: xr.Dataset):
    print(type(data))
    return data
this_ds = xrDataset_function(ekds)
```

```python
@transform_inputs_decorator()
def xr_function(data: xr.Dataset | xr.DataArray):
    print(type(data))
    return data
this_ds = xr_function(ekds)
```

```python
@transform_inputs_decorator()
def np_function(data: np.ndarray):
    print(type(data))
    return data
this_np = np_function(ekds)
```

```python
@transform_inputs_decorator()
def np_str_function(data: str | np.ndarray):
    print(type(data))
    return data

this_np_str = np_str_function(ekds)
this_converted_str = np_str_function("remote_era5_file")
this_converted_str = np_str_function(1)
```
